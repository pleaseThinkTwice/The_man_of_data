```python
def itemsets_length(itemsets):
    l = 0
    for i in itemsets:
        l += len(itemsets[i])
    return l
def rules_lenth(rules):
    return len(rules)
def final_result(go):
    result=[]
    n=100
#     for i in range(1,31):
#         for j in range(1,61):
    for i in range(10,61):
        for j in range(60,101):
    
            itemsets, rules = apriori(go, min_support=i/n, min_confidence=j/n)
            result.append([i/100,j/100,itemsets_length(itemsets),rules_lenth(rules)])
            print(i,end="\t")
            print(j)
    return result
```

```python

itemsets, rules = apriori(go_2, min_support=0.5,  min_confidence=0.88)
print(itemsets_length(itemsets))
print(rules_lenth(rules))
```

```python
result=final_result()
```
